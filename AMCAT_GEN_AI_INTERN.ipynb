# Univarient Analysis

```python
salary_description =df["Salary"].describe()
salary_description.loc['mean'] = round(salary_description['mean'], 2)
salary_description.loc['std'] = round(salary_description['std'], 2)
salary_description.loc['min'] = round(salary_description['min'], 2)
salary_description.loc['25%'] = round(salary_description['25%'], 2)
salary_description.loc['50%'] = round(salary_description['50%'], 2)
salary_description.loc['75%'] = round(salary_description['75%'], 2)
salary_description.loc['max'] = round(salary_description['max'], 2)

print(salary_description)
```

```python
sns.kdeplot(data=df["Salary"])
plt.grid()
plt.title("Distribution of Salary")
plt.show()
```

```python
skewness = df["Salary"].skew()
kurtosis = df["Salary"].kurtosis()
print(f'Skewness: {skewness}, Kurtosis: {kurtosis}')
```

# Distribution of Salary

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=30, kde=True)
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

# Count of Gender

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df)
plt.title('Count of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# Distribution of College GPA

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['collegeGPA'], bins=30, kde=True)
plt.title('Distribution of College GPA')
plt.xlabel('College GPA')
plt.ylabel('Frequency')
plt.show()
```

# Count of Designation

```python
# Get the top 12 designations by count
top_designations = df['Designation'].value_counts().nlargest(12).index

# Filter the DataFrame to include only the top 12 designations
top_df = df[df['Designation'].isin(top_designations)]

# Create the count plot
plt.figure(figsize=(12, 6))
sns.countplot(y='Designation', data=top_df, order=top_designations)
plt.title('Count of Top 12 Designations')
plt.xlabel('Count')
plt.ylabel('Designation')
plt.show()
```

# Distribution of 12th Percentage

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['12percentage'], bins=30, kde=True)
plt.title('Distribution of 12th Percentage')
plt.xlabel('12th Percentage')
plt.ylabel('Frequency')
plt.show()
```

# Bivariate Analysis

# Salary vs. Designation

```python
# Get the top 20 designations by median salary
top_designations = df.groupby('Designation')['Salary'].median().nlargest(20).index

# Filter the DataFrame to include only the top 20 designations
top_df = df[df['Designation'].isin(top_designations)]

# Create the box plot
plt.figure(figsize=(12, 6))
sns.boxplot(y='Designation', x='Salary', data=top_df, order=top_designations)
plt.title('Salary Distribution by Top 20 Designations')
plt.xlabel('Salary')
plt.ylabel('Designation')
plt.show()
```

# 12th Percentage vs. College GPA

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='12percentage', y='collegeGPA', data=df)
plt.title('12th Percentage vs. College GPA')
plt.xlabel('12th Percentage')
plt.ylabel('College GPA')
plt.show()
```

# Count of Gender by Designation

```python
# Get the top 12 designations by count
top_designations = df['Designation'].value_counts().nlargest(12).index

# Filter the DataFrame to include only the top 12 designations
top_df = df[df['Designation'].isin(top_designations)]

# Create the count plot
plt.figure(figsize=(12, 6))
sns.countplot(y='Designation', hue='Gender', data=top_df)
plt.title('Count of Gender by Top 12 Designations')
plt.ylabel('Designation')
plt.xlabel('Count')
plt.legend(title='Gender')
plt.show()
```

# Salary vs. College Tier

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='CollegeTier', y='Salary', data=df)
plt.title('Salary Distribution by College Tier')
plt.xlabel('College Tier')
plt.ylabel('Salary')
plt.show()
```

# FacetGrid for Gender and Designation

```python
# First, let's see how many unique values are in 'Gender' and 'Designation'

# If there are too many categories, you may want to filter or combine them.
# For demonstration, let's assume we only want to look at the first 3 unique values for each.

gender_subset = df['Gender'].unique()[:2]  # Get the first two genders
designation_subset = df['Designation'].unique()[:5]  # Get the first five designations

# Filter the DataFrame for the subsets
filtered_df = df[(df['Gender'].isin(gender_subset)) & (df['Designation'].isin(designation_subset))]

# Create the FacetGrid
g = sns.FacetGrid(filtered_df, col='Gender', row='Designation', margin_titles=True)
g.map(sns.histplot, 'Salary', bins=30, kde=True)
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Salary Distribution by Gender and Designation')
plt.show()
```

# 3D Scatter Plot of Salary, College GPA, and 12th Percentage

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Salary'], df['collegeGPA'], df['12percentage'], c='r', marker='o')
ax.set_xlabel('Salary')
ax.set_ylabel('College GPA')
ax.set_zlabel('12th Percentage')
plt.title('3D Scatter Plot of Salary, College GPA, and 12th Percentage')
plt.show()
```
